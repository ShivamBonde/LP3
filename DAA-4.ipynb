```python
def knapsack():
    n = int(input("Enter number of items: "))

    weights = []
    values = []

    for i in range(n):
        w = int(input(f"Weight of item {i+1}: "))
        v = int(input(f"Value of item {i+1}: "))
        weights.append(w)
        values.append(v)

    W = int(input("\nEnter Knapsack Capacity: "))

    # DP array (1D instead of 2D) â†’ simpler
    dp = [0] * (W + 1)

    for i in range(n):
        for cap in range(W, weights[i] - 1, -1):
            dp[cap] = max(dp[cap], values[i] + dp[cap - weights[i]])

    print("\nMaximum Profit =", dp[W])

knapsack()
```
