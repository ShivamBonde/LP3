```python
# Function to build Huffman Tree and generate codes
def huffman_code(symbols, freq):

    # Create list of pairs (frequency, symbol)
    nodes = list(zip(freq, symbols))

    while len(nodes) > 1:
        # Sort by frequency
        nodes = sorted(nodes)

        # Pick two smallest
        (f1, s1) = nodes[0]
        (f2, s2) = nodes[1]

        # Combine them
        new_node = (f1+f2, s1+s2)

        nodes = nodes[2:]
        nodes.append(new_node)

    return nodes[0][1]  # final combined symbol string


# Generate codes from Tree string
def generate_codes(tree, prefix=""):
    if len(tree) == 1:
        print(tree, "->", prefix)
        return

    # Left = 0, Right = 1
    mid = len(tree) // 2
    generate_codes(tree[:mid], prefix + "0")
    generate_codes(tree[mid:], prefix + "1")


# Main Program
symbols = []
freq = []

n = int(input("Enter number of characters: "))

for i in range(n):
    s = input(f"Enter character {i+1}: ")
    f = int(input(f"Enter frequency of '{s}': "))
    symbols.append(s)
    freq.append(f)

tree = huffman_code(symbols, freq)

print("\nHuffman Codes:")
print("==============")
generate_codes(tree)
```
