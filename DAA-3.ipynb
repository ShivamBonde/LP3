```python
def fractional_knapsack():
    n = int(input("Enter number of items: "))

    items = []
    for i in range(n):
        w = float(input(f"Weight of item {i+1}: "))
        v = float(input(f"Value of item {i+1}: "))
        items.append([w, v, v/w])   # weight, value, ratio

    capacity = float(input("\nEnter knapsack capacity: "))

    # Sort by ratio (value/weight) in descending order
    items.sort(key=lambda x: x[2], reverse=True)

    total = 0
    print("\nTaking items:")

    for w, v, r in items:
        if capacity <= 0:
            break
        
        if w <= capacity:
            total += v
            capacity -= w
            print(f"Full item taken: weight={w}, value={v}")
        else:
            total += r * capacity
            print(f"Partial item taken: weight used={capacity}, value added={r * capacity:.2f}")
            capacity = 0

    print(f"\nMaximum total value = {total:.2f}")

fractional_knapsack()
```
