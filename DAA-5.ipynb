```python
import random

# Deterministic QuickSort
def quick_sort_det(arr, low, high):
    if low < high:
        pivot = partition(arr, low, high)
        quick_sort_det(arr, low, pivot - 1)
        quick_sort_det(arr, pivot + 1, high)

def partition(arr, low, high):
    pivot = arr[high]
    i = low - 1
    for j in range(low, high):
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    return i + 1


# Randomized QuickSort
def quick_sort_rand(arr, low, high):
    if low < high:
        rand_index = random.randint(low, high)
        arr[rand_index], arr[high] = arr[high], arr[rand_index]  # Random pivot
        pivot = partition(arr, low, high)
        quick_sort_rand(arr, low, pivot - 1)
        quick_sort_rand(arr, pivot + 1, high)


# Main
n = int(input("Enter number of elements: "))
arr = list(map(int, input("Enter elements: ").split()))

arr1 = arr.copy()
arr2 = arr.copy()

quick_sort_det(arr1, 0, n - 1)
quick_sort_rand(arr2, 0, n - 1)

print("\nDeterministic QuickSort Output:", arr1)
print("Randomized QuickSort Output:", arr2)
```
